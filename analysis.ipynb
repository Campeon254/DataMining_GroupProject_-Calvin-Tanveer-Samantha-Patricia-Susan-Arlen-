1. The aim of this analysis is to determine whether age is significantly associated with the occurrence of cancer. Specifically, I want to assess whether age may be a risk factor or contributor to cancer development.

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("C:/Users/Admin/OneDrive/Desktop/Lung Cancer.csv")

#Age distribution of lung cancer patients
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=15, kde=True, color='purple')
plt.title('age Distribution of Lung Cancer Patients')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

The histogram show how lung cancer cases are spread across different ages. In this this case, age is a significant factor in lung cancer occurrence, with middle aged individuals at higher risk.

```python
# Smoking status vs Cancer Stage
sns.boxplot(data=df, x='smoking_status', y='age')
plt.title('smoking status vs age')
plt.show()

# Smoking status impact
pd.crosstab(df['smoking_status'], df['cancer_stage']).plot(kind='bar', figsize=(10,6))
plt.title('cancer_stage by smoking status')
plt.ylabel('Number of Patients')
plt.show()
```

a. The boxplot comparing smoking status with age shows that the age distribution is relatively consistent across all smoking categories. The median age and inter quartile ranges are very similar for passive smokers, former smokers, current smokers, and individuals who have never smoked. This suggests that age is not significantly different across smoking status groups, and thus may not be a major confounder in analyzing the impact of smoking on cancer outcomes

b. The bar chart shows the distribution of cancer stages across different smoking statuses. The number of patients in each stage appears nearly identical regardless of smoking status, suggesting no strong visual association between smoking behavior and cancer stage.
